## 1. Prepare Directory Structure for Systematic Alignment of FASTQ Files

* 1.1 Renaming files to reflact sample identity

Directory name was added to each file inside the directory to identify files correctly by the sample it belongs.
Following python script takes input argument of directory name and rename all files within the directory by prefixing the directory name. The script is run with the directory name as argument.

````python
import sys
from os import walk, path, rename

for dirpath, _, files in walk(sys.argv[1]):
    for f in files:
        rename(path.join(dirpath, f), path.join(dirpath, path.split(dirpath)[-1] + '_' + f))
````
